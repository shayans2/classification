```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

from LogisticRegression import LogisticRegression
from utils import plot_2d_boundary, get_line_eq
```

```python
df = pd.read_csv("./Datasets/Data5-moon-2D.csv")
df.head()
```

```python
feats = df.drop('target', axis=1).values
target = df.drop(['Feature 0', 'Feature 1'], axis=1).values.reshape(-1)
```

```python
sns.scatterplot(x=feats[:, 0], y=feats[:, 1], hue=target)
```

```python
X, y = feats, target
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=5)

model = LogisticRegression(learning_rate=0.5, lambda_=0.4, feat_map_deg=3)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print(classification_report(y_pred, y_test))
```

```python
plot_2d_boundary(model, X, y)
get_line_eq(model)
```

### Dataset 2


```python
def predict_and_plot(df, learning_rate=0.1, lambda_=0.4, feat_map_deg=3):
    display(df.head())

    feats = df.drop('target', axis=1).values
    target = df.drop(['Feature 0', 'Feature 1'], axis=1).values.reshape(-1)

    print("Data Shape:")
    sns.scatterplot(x=feats[:, 0], y=feats[:, 1], hue=target)
    plt.show()

    X, y = feats, target
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.2, random_state=5)

    model = LogisticRegression(learning_rate, lambda_, feat_map_deg)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    print(classification_report(y_pred, y_test))

    print("\n Fitted Line:")
    plot_2d_boundary(model, X, y)
    print(get_line_eq(model))
```

```python
df = pd.read_csv("./Datasets/Data2-circles-2D.csv")
predict_and_plot(df)
```
